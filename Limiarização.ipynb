# Limiarização Global

```python
img = cv2.imread('/content/imagens/paisagem01.jpg') # imagem
plt.imshow(img); # ver imagem
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converter para RGB
plt.imshow(rgb); # ver imagem
```

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY) # converter para escala de cinza
plt.imshow(gray, cmap='gray'); # ver imagem
# cmap='gray' : para conseguirmos ver em escala de cinza
```

```python
limiar = 180 # 0 - 255
```

```python
# 255 - valores maximos ddo limiar
# cv2.THRESH_BINARY - limiarização binária
val, thersh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
```

```python
limiar = 100
val, thersh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
plt.imshow(thersh, cmap='gray');
```

```python
limiar = 140 #@param {type: "slider", min: 0, max: 500, step:1}
val, thersh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
plt.imshow(thersh, cmap='gray');
```

```python
def mostrar(imagem):
  fig = plt.gcf() # rejeitar os eixos do grafico
  fig.set_size_inches(18, 6)
  plt.imshow(imagem, cmap='gray');
  plt.axis('off') # retirar os valores nos eixos x e y
  plt.show()
```

```python
mostrar(thersh)
```
