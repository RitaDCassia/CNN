```python
import time
import random
import matplotlib.pyplot as plt
```

1. Implementação dos Algoritmos

Quicksort


```python
# quicksort
"""
Função que implementa o algoritmo Quicksort.

Parâmetros:
  arr (list): lista a ser ordenada
  counters (dict): dicionário para contar comparações e trocas de elementos

Retorno:
  list: lista ordenada
"""
def quicksort(arr, counters):
    if len(arr) <= 1:  # Caso base: se a lista tem 0 ou 1 elemento, já está ordenada
        return arr

    pivot = arr[len(arr) // 2]  # Escolhe o pivô como o elemento do meio

    # Cria sublistas com base no pivô e conta comparações
    left = []
    right = []

    for x in arr:
        counters["comparisons"] += 1  # Conta a comparação
        if x < pivot:
            left.append(x)
        elif x > pivot:
            right.append(x)

    # Conta o número de trocas (neste caso, é uma estimativa)
    counters["swaps"] += len(arr) - len(left) - len(right) - 1

    # Recursão para ordenar as sublistas
    return quicksort(left, counters) + [pivot] + quicksort(right, counters)
```

Heapsort


```python
# Heapsort
def heapsort(arr, counters):
  """
  Função que implementa o algoritimo heapsort.

  Parâmetros:
    arr (list): lista a ser ordenada
    counters (dict): dicionário para contar comparações e trocas de elementos

  Retorno:
    list: lista ordenada
  """
  def heapfy(arr, n, i):
    """
    Função auxiliar para manter a propriedade do heap.

    Parâmetros:
      arr (list): lista a ser organizada
      n (int): tamanho da lista
      i (int): índice do elemento atual a ser verificado
    """

    largest = i # inicia o maior como raiz
    l, r = 2 * i + 1, 2 * i + 2 # calcula os índices dos filhos

    # verifica se o filho esquerdo é maior que a raiz
    if l < n:
      counters["comparisons"] += 1
      if arr[i] < arr[l]:
        largest = l

    # verifica se o filho direito é maior que o maior até o momento
    if r < n:
      counters["comparisons"] += 1
      if arr[r] > arr[largest]:
        largest = r

    # se maior nao for a raiz, troca e continua o heap
    if largest != i:
      arr[i], arr[largest] = arr[largest], arr[i] # troca o maior com a raiz
      counters["swaps"] += 1 # conta a troca
      heapfy(arr, n, largest) # recursao para heapficar a raiz

  n = len(arr)

  # constroi o heapc(reorganiza a lista)
  for i in range(n // 2 - 1, -1, -1):
    heapfy(arr, n, i)

  # extrai elementos do heap um por um
  if i in range(n - 1, 0, -1):
    arr[i], arr[0] = arr[0], arr[i] # troca o maior com a raiz
    counters["swaps"] += 1 # conta a troca
    heapfy(arr, i, 0) # chama heapify na lista reduzida
```

shellsort


```python
# Shellsort
def shellsort(arr, counters):
    """
    Função que implementa o algoritmo Shellsort.

    Parâmetros:
      arr (list): A lista a ser ordenada.
      counters (dict): Dicionário para contar comparações e trocas.
    """
    n = len(arr)
    gap = n // 2  # Inicializa o intervalo (gap)

    # Continua até que o gap seja 0
    while gap > 0:
        # Faz uma inserção em cada sublista
        for i in range(gap, n):
            temp = arr[i]  # Armazena o valor atual
            j = i

            # Move os elementos que estão no intervalo de gap para a direita
            while j >= gap:
                counters["comparisons"] += 1
                if arr[j - gap] > temp:
                    arr[j] = arr[j - gap]
                    j -= gap
                    counters["swaps"] += 1  # Conta a troca
                else:
                    break  # Se não for maior, não precisa continuar

            arr[j] = temp  # Coloca o valor armazenado na posição correta
        gap //= 2  # Reduz o intervalo
```

3.  Execução e Medição

```python

# Função para medir desempenho
def execute_and_measure(algorithm, data):
    """
    Executa o algoritmo e mede o tempo, comparações e trocas.

    Parâmetros:
        algorithm (function): Função do algoritmo de ordenação.
        data (list): Dados a serem ordenados.

    Retorno:
        tuple: Tempo de execução, contadores de comparações e trocas.
    """
    counters = {"comparisons": 0, "swaps": 0}
    start_time = time.time()
    algorithm(data.copy(), counters)
    elapsed_time = time.time() - start_time
    return elapsed_time, counters
```

```python
# Cenários de teste
sizes = [10, 157, 3450, 10000, 987890]
test_cases = {
    "Random": lambda size: [random.randint(1, 1000) for _ in range(size)],
    "Sorted": lambda size: list(range(size)),
    "Reverse": lambda size: list(range(size, 0, -1)),
    "Almost Sorted": lambda size: sorted([random.randint(1, size) for _ in range(size)]),
}
```

```python
# Executar testes
results = {algo: {case: [] for case in test_cases} for algo in ["Quicksort", "Heapsort", "Shellsort"]}

for size in sizes:
    for case_name, case_gen in test_cases.items():
        data = case_gen(size)
        for algo_name, algo in [("Quicksort", quicksort), ("Heapsort", heapsort), ("Shellsort", shellsort)]:
            time_taken, counters = execute_and_measure(algo, data)
            results[algo_name][case_name].append({
                "size": size,
                "time": time_taken,
                "comparisons": counters["comparisons"],
                "swaps": counters["swaps"],
            })
```

```python
# Exibir resultados
for algo, cases in results.items():
    print(f"\n{algo} Resultado:")
    for case_name, case_results in cases.items():
        print(f"  {case_name}:")
        for result in case_results:
            print(f"    tamanho: {result['size']}, Tempo: {result['time']:.6f}s, "
                  f"Comparações: {result['comparisons']}, Swaps: {result['swaps']}")
```

4. Análise e Relatório

```python
plt.figure(figsize=(12, 6))
for algo_name in results:
    for case_name, metrics in results[algo_name].items():
        times = [m["time"] for m in metrics]
        plt.plot(sizes, times, marker='o', label=f"{algo_name} ({case_name})")
plt.xlabel("Tamanho da Entrada")
plt.ylabel("Tempo de Execução (s)")
plt.title("Algoritmos de Ordenação - Tempo de Execução")
plt.legend()
plt.grid()
```

```python
# Gráficos - Comparações
plt.figure(figsize=(12, 6))  # Cria uma nova figura com tamanho 12x6
for algo_name in results:  # Itera sobre os nomes dos algoritmos nos resultados
    for case_name, metrics in results[algo_name].items():  # Itera sobre os casos de teste e suas métricas
        comparisons = [m["comparisons"] for m in metrics]  # Extrai o número de comparações para cada métrica
        plt.plot(sizes, comparisons, marker='o', label=f"{algo_name} ({case_name})")  # Plota o número de comparações
plt.xlabel("Tamanho da Entrada")  # Rótulo do eixo X
plt.ylabel("Número de Comparações")  # Rótulo do eixo Y
plt.title("Algoritmos de Ordenação - Comparações")  # Título do gráfico
plt.legend()  # Adiciona a legenda
plt.grid()  # Adiciona uma grade ao gráfico
plt.show()  # Exibe o gráfico
```

```python
# Gráficos - Trocas
plt.figure(figsize=(12, 6))  # Cria uma nova figura com tamanho 12x6
for algo_name in results:  # Itera sobre os nomes dos algoritmos nos resultados
    for case_name, metrics in results[algo_name].items():  # Itera sobre os casos de teste e suas métricas
        swaps = [m["swaps"] for m in metrics]  # Extrai o número de trocas para cada métrica
        plt.plot(sizes, swaps, marker='o', label=f"{algo_name} ({case_name})")  # Plota o número de trocas
plt.xlabel("Tamanho da Entrada")  # Rótulo do eixo X
plt.ylabel("Número de Trocas")  # Rótulo do eixo Y
plt.title("Algoritmos de Ordenação - Trocas")  # Título do gráfico
plt.legend()  # Adiciona a legenda
plt.grid()  # Adiciona uma grade ao gráfico
plt.show()  # Exibe o gráfico
```
