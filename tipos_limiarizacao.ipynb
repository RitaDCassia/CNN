```python
def exibir_limiarizacao(img, limiar = 127):
  _, thresh_binary = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY) # binarizacao
  _, thresh_binary_inv = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY_INV) # binarizacao
  _, thresh_trunc = cv2.threshold(gray, limiar, 255, cv2.THRESH_TRUNC) # binarizacao
  _, thresh_to_zero = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY) # binarizacao
  _, thresh_to_zero_inv = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY_INV) # binarizacao

  titulos = ['Imagem Original', 'Binary', 'Binary Inverted', 'Trunc', 'To Zero', 'To Zero Inverted']

  imagens = [img, thresh_binary, thresh_binary_inv, thresh_trunc, thresh_to_zero, thresh_to_zero_inv]

  fig = plt.gcf()
  fig.set_size_inches(18, 12)

  for i in range(6):
    plt.subplot(2, 3, i+1) # numero de linhas e colunas e id de imagens
    plt.imshow(cv2.cvtColor(imagens[i], cv2.COLOR_GRAY2RGB), cmap='gray')
    plt.title(titulos[i])
    plt.xticks([])
    plt.yticks([])
  plt.show()
```

```python
exibir_limiarizacao(gray, limiar)
```
