```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def buscaBinaria(arr, item, inicio, fim):
    """
    Implementa uma busca binária para encontrar a posição de inserção de um item em um array ordenado.

    Args:
        arr (list): Array ordenado onde será realizada a busca
        item: Elemento a ser inserido
        inicio (int): Índice inicial do intervalo de busca
        fim (int): Índice final do intervalo de busca

    Returns:
        int: Posição onde o item deve ser inserido para manter o array ordenado
    """

    # Caso base: quando o intervalo de busca se reduz a um elemento ou menos
    if fim <= inicio:
        # Se o item for maior que o elemento na posição inicial,
        # deve ser inserido após ele
        if item > arr[inicio]:
            return inicio + 1
        # Caso contrário, deve ser inserido na posição atual
        return inicio

    # Calcula o ponto médio do intervalo atual
    meio = (inicio + fim) // 2

    # Compara o item com o elemento do meio do intervalo
    if item == arr[meio]:
        # Se for igual, retorna a posição após o elemento encontrado
        return meio + 1
    elif item > arr[meio]:
        # Se for maior, busca na metade direita do intervalo
        # Recursivamente chama a busca do meio+1 até o fim
        return buscaBinaria(arr, item, meio + 1, fim)
    else:
        # Se for menor, busca na metade esquerda do intervalo
        # Recursivamente chama a busca do inicio até o meio-1
        return buscaBinaria(arr, item, inicio, meio - 1)
```

```python
def insertion_sort_tradicional(arr):
    """
    Implementa o algoritmo de ordenação Insertion Sort tradicional,
    contabilizando o número de comparações e movimentações realizadas.

    Args:
        arr (list): Lista a ser ordenada

    Returns:
        tuple: (comparacoes, movimentacoes)
            - comparacoes (int): Número total de comparações realizadas
            - movimentacoes (int): Número total de movimentações realizadas

    Complexidade:
        - Tempo: O(n²)
        - Espaço: O(1)
    """

    # Inicializa contadores de operações
    comparacoes = 0    # Conta número de comparações
    movimentacoes = 0  # Conta número de movimentações

    # Percorre a lista a partir do segundo elemento
    for i in range(1, len(arr)):
        # Seleciona o elemento atual para inserção
        chave = arr[i]
        # Inicializa o índice do elemento anterior
        j = i - 1

        # Move elementos maiores que a chave uma posição à frente
        while j >= 0 and arr[j] > chave:
            comparacoes += 1     # Incrementa contador de comparações
            movimentacoes += 1   # Incrementa contador de movimentações
            arr[j + 1] = arr[j]  # Move elemento para frente
            j -= 1              # Move para o próximo elemento a comparar

        # Insere a chave na posição correta
        arr[j + 1] = chave

    return comparacoes, movimentacoes
```

```python
def insertion_sort_binario(arr):
    """
    Implementa o algoritmo Insertion Sort modificado com busca binária,
    contabilizando o número de comparações e movimentações realizadas.

    Esta versão utiliza busca binária para encontrar a posição de inserção,
    reduzindo o número de comparações de O(n²) para O(n log n).

    Args:
        arr (list): Lista a ser ordenada

    Returns:
        tuple: (comparacoes, movimentacoes)
            - comparacoes (int): Número total de comparações (O(n log n))
            - movimentacoes (int): Número total de movimentações (O(n²))

    Complexidade:
        - Tempo total: O(n²) devido às movimentações
        - Comparações: O(n log n)
        - Espaço: O(1)
    """

    # Inicializa contadores de operações
    comparacoes = 0    # Conta número de comparações
    movimentacoes = 0  # Conta número de movimentações

    # Percorre a lista a partir do segundo elemento
    for i in range(1, len(arr)):
        # Seleciona o elemento atual para inserção
        chave = arr[i]
        j = i - 1

        # Usa busca binária para encontrar a posição de inserção
        # A busca binária reduz as comparações para log n
        pos = buscaBinaria(arr, chave, 0, j)
        # Adiciona log2(i) comparações para cada elemento i
        comparacoes += int(np.log2(i)) if i > 0 else 0

        # Move os elementos para abrir espaço para inserção
        # Esta parte ainda requer O(n) movimentações no pior caso
        while j >= pos:
            movimentacoes += 1   # Conta cada movimentação
            arr[j + 1] = arr[j]  # Move elemento uma posição à frente
            j -= 1

        # Insere a chave na posição encontrada pela busca binária
        arr[pos] = chave

    return comparacoes, movimentacoes
```

```python
def comparar_algoritmos():
    """
    Realiza uma análise comparativa entre o Insertion Sort tradicional e sua versão
    modificada com busca binária, gerando gráficos e métricas de desempenho.

    A função:
    1. Testa diferentes tamanhos de entrada
    2. Mede número de comparações e movimentações
    3. Gera gráficos comparativos
    4. Imprime estatísticas numéricas

    Requer:
        - numpy (np)
        - matplotlib.pyplot (plt)

    Complexidade:
        - Tempo: O(n²) para cada teste
        - Espaço: O(n) para arrays de teste
    """

    # Define tamanhos de entrada para teste
    n = np.array([10, 50, 100, 500, 1000, 2000])

    # Inicializa arrays para armazenar resultados
    comp_trad = []  # Comparações do método tradicional
    comp_bin = []   # Comparações do método com busca binária
    mov_trad = []   # Movimentações do método tradicional
    mov_bin = []    # Movimentações do método com busca binária

    # Realiza testes para cada tamanho de entrada
    for size in n:
        # Gera array aleatório para teste
        arr = list(np.random.randint(1, 1000, size))

        # Testa versão tradicional
        c1, m1 = insertion_sort_tradicional(arr.copy())
        comp_trad.append(c1)
        mov_trad.append(m1)

        # Testa versão com busca binária
        c2, m2 = insertion_sort_binario(arr.copy())
        comp_bin.append(c2)
        mov_bin.append(m2)

    # Configuração do plot
    plt.figure(figsize=(15, 6))

    # Gráfico de comparações
    plt.subplot(1, 2, 1)
    plt.plot(n, comp_trad, 'r-', label='Tradicional (O(n²))', linewidth=2)
    plt.plot(n, comp_bin, 'b-', label='Busca Binária (O(n log n))', linewidth=2)
    plt.plot(n, n*np.log2(n), 'g--', label='n log n', linewidth=1)
    plt.plot(n, n**2/10, 'r--', label='n²', linewidth=1)
    plt.title('Número de Comparações', fontsize=12)
    plt.xlabel('Tamanho da Entrada (n)', fontsize=10)
    plt.ylabel('Número de Comparações', fontsize=10)
    plt.legend()
    plt.grid(True)

    # Gráfico de movimentações
    plt.subplot(1, 2, 2)
    plt.plot(n, mov_trad, 'r-', label='Tradicional', linewidth=2)
    plt.plot(n, mov_bin, 'b-', label='Busca Binária', linewidth=2)
    plt.plot(n, n**2/10, 'r--', label='n²', linewidth=1)
    plt.title('Número de Movimentações', fontsize=12)
    plt.xlabel('Tamanho da Entrada (n)', fontsize=10)
    plt.ylabel('Número de Movimentações', fontsize=10)
    plt.legend()
    plt.grid(True)

    # Configurações finais do plot
    plt.suptitle('Comparação entre Insertion Sort Tradicional e com Busca Binária',
                fontsize=14)
    plt.tight_layout()
    plt.show()

    # Imprime estatísticas para o maior tamanho testado
    print("\nComparação numérica para n=2000:")
    print(f"Comparações (Tradicional): {comp_trad[-1]}")
    print(f"Comparações (Busca Binária): {comp_bin[-1]}")
    print(f"Movimentações (Tradicional): {mov_trad[-1]}")
    print(f"Movimentações (Busca Binária): {mov_bin[-1]}")
```

```python
# Executar a comparação
comparar_algoritmos()
```
